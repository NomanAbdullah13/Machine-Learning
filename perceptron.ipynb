```python
lr = 0.5
theta = 0.5
```

```python
count = 0
epoch = 0
loss = []
n = len(y)

while count < n:
  epoch += 1
  total_error = 0
  print(f"Epoch: {epoch}")

  for i in range(n):
    wx = w[0] * X[i][0] + w[1] * X[i][1] + b
    y_pred = 1 if wx >= theta else 0
    error = y[i] - y_pred
    if error != 0:
      w[0] += lr * error * X[i][0]
      w[1] += lr * error * X[i][1]
      b += lr * error
      total_error += error
      count = 0
    else:
      count += 1
  loss.append(np.abs(total_error / n))

  if w[1] != 0:
    x_intercept = -b / w[0]
    y_intercept = -b / w[1]
    slope = -w[0] / w[1]
  else:
    x_intercept = 0
    y_intercept = b
    slope = float('inf')

  x_values = np.linspace(0, max(X[:, 0]), 100)
  y_values = slope * x_values + y_intercept

  display.clear_output(wait=True)

  plt.figure(figsize=(20, 5))

  plt.subplot(1, 2, 1)
  colors = ['red' if label == 0 else 'green' for label in y]
  plt.scatter(X[:, 0], X[:, 1], c=colors)
  plt.plot(x_values, y_values, color='blue', label='Decision Boundary')
  plt.xlabel('X1')
  plt.ylabel('X2')
  plt.title('Data Points and Decision Boundary')

  plt.subplot(1, 2, 2)
  plt.plot(loss, label='Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.title('Loss vs Epoch')

  plt.show()
```
